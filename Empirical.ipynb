```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('BTC-USD_1.csv', index_col='Date')
df.tail()
```

```python
market = df["Adj Close"]
market
```

```python
returns = 100 * market.pct_change().dropna()
returns
```

```python
returns.plot(figsize=(10,12), subplots=True)
plt.ylabel('BTC daily returns')
plt.xlabel('Date')
```
